```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
df=pd.read_csv('reduced_alzheimers_disease_data.csv')
print(df)
```

```python
x=df[["Age","Gender","EducationLevel","Smoking","PhysicalActivity","DietQuality",
      "DifficultyCompletingTasks","PersonalityChanges","Disorientation","Confusion",
      "BehavioralProblems","MemoryComplaints","FunctionalAssessment","CholesterolTotal","DiastolicBP",
      "SystolicBP","Hypertension","HeadInjury","ADL","MMSE","Depression","SleepQuality",
      "FamilyHistoryAlzheimers","CardiovascularDisease","Diabetes"]]

y=df["Diagnosis"]
x= pd.get_dummies(x)
```

```python
# train-test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#1.RANDOM FOREST CLASSIFIER
```

```python
model1= RandomForestClassifier()
model1.fit(x_train, y_train)
y_prediction = model1.predict(x_test)
accuracy = accuracy_score(y_test, y_prediction)
print("RandomForestClassifier Accuracy:", accuracy*100)
joblib.dump(model1,"mymodel_ran.h5")
# Load and predict with Random Forest
loaded_model1 = joblib.load("mymodel_ran.h5")
new_data = x_test.iloc[:1]
rf_predictions = loaded_model1.predict(new_data)
print("Random Forest Predictions on new data:", rf_predictions)
```

```python
model3 = LogisticRegression()
model3.fit(x_train, y_train)
y_prediction = model3.predict(x_test)
accuracy = accuracy_score(y_test, y_prediction)
print("LogisticRegression Accuracy:", accuracy*100)
joblib.dump(model3,"mymodel_log.h5")
#load and prediction
loaded_model3 = joblib.load("mymodel_log.h5")
new_data = x_test.iloc[:5]  
# Just an example, use actual new data here
log_reg_predictions = loaded_model3.predict(new_data)
print("Logistic Regression Predictions on new data:", log_reg_predictions)
```

```python
model4 = SVC()
model4.fit(x_train, y_train)
y_prediction = model4.predict(x_test)
accuracy = accuracy_score(y_test, y_prediction)
print("SVC Accuracy:", accuracy*100)
joblib.dump(model4,"mymodel_sv.h5")
```

```python
import pandas as pd
import joblib

# Load the model (ensure it's a .pkl file, not a .csv)
model = joblib.load("mymodel_ran.h5")


# Get feature names from the model
feature_names = model.feature_names_in_

# Input data from the user
try:
    a = int(input("Enter your age: "))
    b = float(input("Enter your Gender: "))
    c = int(input("Enter Education Level: "))
    d = int(input("Enter if you smoke (0, 1): "))
    e = float(input("Enter if you consume alcohol (0-10 scale): "))
    f = float(input("What is your Physical Activity (0-10 scale): "))
    g = float(input("How is your diet quality (0-10 scale): "))
    h = float(input("How is your sleep quality (0-10 scale): "))
    i = int(input("Did your family have any Alzheimers (0, 1): "))
    j = int(input("Did your family have any Cardiovascular Disease (0, 1): "))
    k = int(input("Did you have Diabetes (0, 1): "))
    l = int(input("Are you in depression (0, 1): "))
    m = int(input("Did you have any head injury (0, 1): "))
    n = int(input("Did you have hypertension (0, 1): "))
    o = int(input("Did you have systolic BP (0, 1): "))
    p = int(input("Did you have diastolic BP (0, 1): "))
    q = float(input("How much is your Cholesterol (in mg/dL): "))
    r = float(input("What is your MMSE (Mini-Mental State Examination) score: "))
    s = float(input("What is your functional assessment score: "))
    t = int(input("Do you have any memory complaints (0, 1): "))
    u = int(input("Do you have any behavioral problems (0, 1): "))
    v = int(input("Do you have any ADL (Activities of Daily Living) difficulties (0, 1): "))
    w = int(input("Are you a confused person (0, 1): "))
    x = int(input("Do you have any disorientation (0, 1): "))
    y = int(input("Did you notice any personality changes (0, 1): "))
    z = int(input("Do you have any problems in completing tasks (0, 1): "))
    a1 = int(input("Do you have forgetfulness (0, 1): "))

    # Prepare the data for prediction
    new_data = {
        "Age": a,
        "Gender": b,
        "EducationLevel": c,
        "Smoking": d,
        "AlcoholConsumption": e,
        "PhysicalActivity": f,
        "DietQuality": g,
        "SleepQuality": h,
        "FamilyHistoryAlzheimers": i,
        "CardiovascularDisease": j,
        "Diabetes": k,
        "Depression": l,
        "HeadInjury": m,
        "Hypertension": n,
        "SystolicBP": o,
        "DiastolicBP": p,
        "CholesterolTotal": q,
        "MMSE": r,
        "FunctionalAssessment": s,
        "MemoryComplaints": t,
        "BehavioralProblems": u,
        "ADL": v,
        "Confusion": w,
        "Disorientation": x,
        "PersonalityChanges": y,
        "DifficultyCompletingTasks": z,
        "Forgetfulness": a1,
    }

    # Convert the data to a DataFrame
    new_data_df = pd.DataFrame([new_data])

    # Ensure the features align with the model's input
    new_data_encoded = new_data_df.reindex(columns=feature_names, fill_value=0)

    # Make the prediction
    prediction = model.predict(new_data_encoded)
    print("Prediction result: ", prediction)

except ValueError as e:
    print(f"Error: Invalid input. Please enter valid data. {e}")
```
